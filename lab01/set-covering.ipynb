```python
from random import random
from functools import reduce
from collections import namedtuple
from queue import PriorityQueue, SimpleQueue, LifoQueue

import numpy as np
```

```python
PROBLEM_SIZE = 20
NUM_SETS = 40
SETS = tuple(
    np.array([random() < 0.3 for _ in range(PROBLEM_SIZE)])
    for _ in range(NUM_SETS)
)
State = namedtuple('State', ['taken', 'not_taken'])
for i in range(NUM_SETS):
    print(SETS[i])
```

```python
def goal_check(state):
    return np.all(reduce(
        np.logical_or,
        [SETS[i] for i in state.taken],
        np.array([False for _ in range(PROBLEM_SIZE)]),
    ))


def distance(state):
    return PROBLEM_SIZE - sum(
        reduce(
            np.logical_or,
            [SETS[i] for i in state.taken],
            np.array([False for _ in range(PROBLEM_SIZE)]),
        ))

def f(new_state):
    return distance(new_state) + len(new_state.taken)
```

```python
assert goal_check(
    State(set(range(NUM_SETS)), set())
), "Probelm not solvable"
```

# My A* implementation

```python
frontier = PriorityQueue()
# frontier = LifoQueue()
state = State(set(), set(range(NUM_SETS)))
frontier.put((f(state), state))

# voglio il minimo numero di set il cui or copre tutti gli elementi
# quindi:  ogni volta che aggiungo un set puo avvicinarmi alla soluzione ma potrebbe non essere quella ottimale
# ovvero "minimo numero di set"

counter = 0
d_curr = PROBLEM_SIZE
_, current_state = frontier.get()
while not goal_check(current_state):
    counter += 1
    for action in current_state[1]:
        # A* : computes the path with minimum cost expanding a minimum number of nodes
        new_state = State(
            current_state.taken ^ {action},
            current_state.not_taken ^ {action},
        )
        print(f"{distance(new_state)} | {d_curr} => {action} {'mi avvicina' if distance(new_state) < d_curr else 'non mi avvicina'}")
        # metto in coda di esplorazione solo quelli che mi avvicinano a GOAL (ovvero se la distanza diminuisce)
        if f(new_state) < d_curr:
            frontier.put((f(new_state), new_state))
    print("____________________________________")
    d_curr, current_state = frontier.get()
    print(f"Estraggo quello che mi avvicina di pi√π: {current_state}")

print(
    f"Solved in {counter:,} steps ({len(current_state.taken)} tiles)"
)
```
